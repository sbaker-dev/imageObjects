# Examples for imageObjects

This work sheet is designed to show all the possible examples of how this process can work to achieve the results you
want semi-interactively. We will be using an example image in this case, but you can download this notebook or change 
it to read an image that you are working with to test the functionality. 

First we need to import the libraries we are going to be using

```python
from imageObjects.ImageObject import ImageObject
from imageObjects import image_maker as im
import cv2
```

### Example setting and inverting
Setting an image as an ImageObject is mainly done one of two ways. If you haven't loaded the image into memory yet, 
then you can use cv2 to load it via imread. Otherwise if you already have an image you can set any cv2 image to be a new
instance of an ImageObject. Lets us load in our example via cv2 imread and then show it with .notebook_show.Keep in mind
that ImageObject also has show, but this is designed for working in .py files, as this was it's core intend use. This 
will likely lead to problems though if you use .show in jupyter and **may cause crashes**, so remember to use 
notebook_show!

Our example below shows us loading the example image in from our repository via cv2.imread as our first example image.
We then have two things we can when we want to perform an operation. If we want to change this image, then we can call
a function such as invert_image() that will over-ride the image stored in that instance of ImageObject. This is shown in
our second image, where now when we ask the same instance of ImageObject called 'image' to be displaced yet we get a 
different image. If we want to create a new instance, then we can set the new_image key word to be true. This will 
return a new instance of ImageObjects and not alter the image in memory. The third image shown now shows the original as
we inverted an inverted image, but if you where to call image.notebook show again you would still have the second image.

If you are ever confused as to what the object contains, then you can always print it and it will give you a set of 
information based on its shape

```python
image = ImageObject(cv2.imread(r"C:\Users\Samuel\PycharmProjects\imageObjects\Examples\Page1.png"))
image.notebook_show("Example image")

image.invert_image()
image.notebook_show("Change Original")

invert = image.invert_image(new_image=True)
invert.notebook_show("New from existing")

print(f"Our ImageObject instance of image has the following properties\n{image}")
```

### Creating a blank

It is also possible that you just want to create a blank image with these properties. You can do that by passing None to
the ImageObject and then calling create_blank. **Keep in mind!** that most functions will not work if you pass a None 
type and don't update the image before calling its properties or functions.

```python
constructed_blank = im.create_blank(50, 50)
print(constructed_blank)
constructed_blank.notebook_show()
```
